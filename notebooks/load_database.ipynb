```python
import pandas as pd
from sqlalchemy import create_engine

print("Everything ready")
```

```python
engine = create_engine("sqlite:///../bluestock_mf.db")

engine
```

```python
nav = pd.read_csv("../data/processed/nav_history_cleaned.csv")

nav.head()
```

```python
nav.to_sql(
    "fact_nav",
    engine,
    if_exists="replace",
    index=False
)
```

```python
pd.read_sql(
    "SELECT COUNT(*) as rows FROM fact_nav",
    engine
)
```
